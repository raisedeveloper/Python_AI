```python
# 데이터프레임 생성
import pandas as pd
data = {'이름': ['Alice', 'Bob', 'Charlie'], '나이': [25, 30, 22], '성별': ['여', '남', '남']}
df = pd.DataFrame(data, index=['A', 'B', 'C'])
print(df)
```

```python
# 첫번째  행 선택 (인덱스 0)
first_row = df.iloc[0]
print(first_row)

# 두 번째 열 (나이) 선택
age_column = df.iloc[:, 1]
print(age_column)

# 첫 번째 행부터 두 번째 행까지, 모든 열 선택
subset = df.iloc[0:2, :]
print(subset)
```

```python
# '이름 열 선택
name_column = df.loc[:, '이름']
print(name_column)

# 'A'행 선택
first_row = df.loc['A']
print(first_row)

# 나이가 25 이상인 행 선택
selected_row = df.loc[df['나이'] >= 25]
print(selected_row)
```
